```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris.keys()
```

```python
iris_data = iris.data
iris_data
```

```python
iris_label = iris.target
iris_label
```

```python
print(iris['feature_names'])
print(iris['target_names'])
```

```python
import pandas as pd
df_iris = pd.DataFrame( data = iris_data, columns= iris.feature_names)
```

```python
iris_copy = df_iris
```

```python
iris_copy.isnull()
```

```python
iris_copy.isnull().sum()
```

```python
iris_copy['species']=iris.target
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# boxplot(x='sepal length (cm)')
plt.figure(figsize = (12,6))
sns.boxplot(x='sepal length (cm)', y='species', data=iris_copy ,orient='h');

# boxplot(x='sepal width (cm)')
plt.figure(figsize = (12,6))
sns.boxplot(x='sepal width (cm)', y='species', data=iris_copy, orient='h');

# boxplot(x='petal length (cm)')
plt.figure(figsize = (12,6))
sns.boxplot(x='petal length (cm)', y='species', data=iris_copy, orient='h');

# boxplot(x='petal width (cm)')
plt.figure(figsize = (12,6))
sns.boxplot(x='petal width (cm)', y='species', data=iris_copy, orient='h');

# pairplot
sns.pairplot(iris_copy, hue='species');
```
