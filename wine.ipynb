```python
from sklearn.datasets import load_wine
import pandas as pd
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
wine_data = load_wine()
wine_target = wine_data.target #와인 데이터에서 target 데이터를 갖고온다.
df_features = wine_data.feature_names # 와인 데이터에서 feature로 되어있는 데이터를 가져온다.
```

```python
df_data = pd.DataFrame(wine_data.data, columns = df_features)
df_label = pd.DataFrame(wine_data.target, columns = ['label'])
data=pd.concat([df_data,df_label],axis=1)
```

```python
sns.countplot(data = data, x="label")
plt.xlabel("class")
plt.ylabel("Count")
plt.title("Label")
```

```python
plt.scatter(data['label'], data['hue'])
plt.xlabel('class')
plt.ylabel('hue')

plt.show()
```

```python
plt.scatter(data['label'], data['color_intensity'])
plt.xlabel('class')
plt.ylabel('color_intensity')

plt.show()
```
